```python
frutas = ['maçã', 'banana', 'uva']
newlist = []

for fruta in frutas:
    if 'b' in fruta:
        newlist.append(fruta)
print(newlist)
```

```python
newlist = [fruta for fruta in frutas if 'b' in fruta]
print(newlist)
```

## Sintaxe básica:
    newlist = [expression for item in iterable if condition == True]

    # expression pode assumir quaisquer operações envolvendo a variável item

```python
numeros = [x for x in range(1, 11)]
squares = [x**2 for x in numeros]

print(numeros)
print(squares)
```

```python
parimpar = [0 if i % 2 == 0 else 1  for i in range(15)]
print(parimpar)
```
